## Odgovori na vprašanja
### Kako ste z analizo ocenili število vrtljajev ventilatorja?
O: Najprej sem pridobil BMF (Blade Pass Frequency), ki sem ga določil kot frekvenco z najvišjo amplitudo v DFT. Nato sem RPM (Rotations Per Minute) izračunal po naslednji formuli:  rpm = (f_bmp * 60) / blade_count.

### Ali so spremembe frekvenčne vsebine linearne glede na hitrost vrtenja pri signalih sig_c_x.
O: Iz grafa je razvidno da je frekvenčna vsebina linearno odvisna od hitrosti vrtenja.

```python
plt.plot(c_peak_freqs, c_rpms, marker='o')
plt.xlabel("Najvišja frekvenca [Hz]")
plt.ylabel("RPM")
plt.show()
```
